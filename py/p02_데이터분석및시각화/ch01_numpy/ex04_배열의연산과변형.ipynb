## 주요 산술 연산 함수
### 1. 수학 함수들

```python
# 배열 생성
# arr = array([1, 2, 3, 4])

# 제곱
print(np.square(arr))

# 제곱근(루트)
print(np.sqrt(arr))

# exp(지수승)
print(np.exp(arr))

# log(로그)
print(np.log(arr))
```

### 2. 배열 연산 함수들

```python
# 더하기
print(np.add(arr, arr1))
print(arr + arr1)

# 합계
print(np.sum(arr))

# 누적합(cumulative + sum)
print(np.cumsum(arr))
```

### 3. 통계 함수들

```python
# 평균
'''
(1 + 2 + 3 + 4)/4 -> 2.5
'''
print(np.mean(arr))
# 분산(variance)
'''
평균: 2.5
편차: -1.5, -0.5, 0.5, 1.5
편차의 제곱: 2.25, 0.25, 0.25, 2.25
분산: (2.25+0.25+0.25+2.25)/4 -> 1.25
'''
print(np.var(arr))
# 표준편차(Standard Deviation)
# 표준편차 -> 분산의 제곱근: 루트1.25 -> 1.118033988749895
print(np.std(arr))

# 최솟값
print(np.min(arr))

# 최댓값
print(np.max(arr))

# 최솟값의 인덱스
print(np.argmin(arr))

# 최댓값의 인덱스
print(np.argmax(arr))
```

## 배열의 변형(Transformation)
### 1. 축 변환: 배열.transpose()

```python
# 2차원 배열 생성
arr = np.arange(10).reshape(2, 5)
print(arr)

# 2차원 배열의 축 변환
# (2, 5) -> (5, 2)
print(arr.transpose())
```
