## 주요 산술 연산 함수
### 1. 수학 함수들

```python
# 배열 생성
# arr = array([1, 2, 3, 4])

# 제곱
print(np.square(arr))

# 제곱근(루트)
print(np.sqrt(arr))

# exp(지수승)
print(np.exp(arr))

# log(로그)
print(np.log(arr))
```

### 2. 배열 연산 함수들

```python
# 더하기
print(np.add(arr, arr1))
print(arr + arr1)

# 합계
print(np.sum(arr))

# 누적합(cumulative + sum)
print(np.cumsum(arr))
```

### 3. 통계 함수들

```python
# 평균
'''
(1 + 2 + 3 + 4)/4 -> 2.5
'''
print(np.mean(arr))
# 분산(variance)
'''
평균: 2.5
편차: -1.5, -0.5, 0.5, 1.5
편차의 제곱: 2.25, 0.25, 0.25, 2.25
분산: (2.25+0.25+0.25+2.25)/4 -> 1.25
'''
print(np.var(arr))
# 표준편차(Standard Deviation)
# 표준편차 -> 분산의 제곱근: 루트1.25 -> 1.118033988749895
print(np.std(arr))

# 최솟값
print(np.min(arr))

# 최댓값
print(np.max(arr))

# 최솟값의 인덱스
print(np.argmin(arr))

# 최댓값의 인덱스
print(np.argmax(arr))
```

## 배열의 변형(Transformation)
### 1. 축 변환: 배열.transpose()

```python
# 2차원 배열 생성
arr = np.arange(10).reshape(2, 5)
print(arr)
```

```python
# 2차원 배열의 축 변환
# (2, 5) -> (5, 2)
print(arr.transpose())
```

```python
# 3차원 배열 생성
# reshape(면, 행, 열)
arr3d = np.arange(27).reshape(3, 3, 3)
print(arr3d)
```

```python
# 3차원 배열의 축 변환
# (0, 1, 2) -> (1, 0, 2)
# 1: 1번 축에있던 '행'을 새로운 배열의 0번 축(면)으로
# 0: 0번 축에있던 '면'을 새로운 배열의 1번 축(행)으로
# 2: 2번 축에있던 '열'은 변함없이 그대로 '2번 축(열)로
print(arr3d.transpose(1, 0, 2))
```

## 고급 연산 기법
### 브로드캐스팅?
- shape(모양)가 다른 다차원배열 또는 스칼라 값 사이에서 연산이 가능하게 만들어 주는 기능

```python
# 2차원 배열과 1차원 배열의 연산
arr_2d = np.array([[1,2,3], [4,5,6]])
arr_1d = np.array([10,20,30])

result = arr_2d + arr_1d
print(result)
```

### 축 지정 연산

```python
# 축 지정 연산
# 열 방향 합계 (axis=0)
print(np.sum(arr_2d, axis=0))

# 행 방향 합계 (axis=1)
print(np.sum(arr_2d, axis=1))

# 전체 합계
print(np.sum(arr_2d))
```

### 조건부 연산

```python
# 배열 생성
arr = np.array([1, -2, 3, -4, 5])

# 절댓값: 양수화
print(np.abs(arr))

# 조건부 연산: np.where(조건식, 참, 거짓)
result = np.where(arr > 0, arr, 0)
print(result)
```

```python
## 실무 활용 예시
### 정규화(Normalization)
```

```python
# 배열 생성
data = np.array([10,20,30,40,50])

# (data - min) / (max - min) 정규화
'''
([10,20,30,40,50] - 10) / (50 - 10)
([10,20,30,40,50] - 10) / 40
[0,10,20,30,40] / 40
[0.,0.25,0.5,0.75,1.]
'''
normalized = (data - np.min(data)) / (np.max(data) - np.min(data))
print(normalized)
```

```python
# z-score 정규화
z_score = (data - np.mean(data)) / np.std(data)
print(z_score)
```

### 이동평균 계산

```python
# 배열 생성
prices = np.array([100, 102, 98, 105, 103, 99, 101])

# 3일 이동평균 함수 생성
# np.convolve(배열, 필터, mode): 합성곱 -> 곱의 합
# np.ones(window)/window
# np.ones(3)/3
# [1.,1.,1.]/3
# [1/3,1/3,1/3]
# [0.33333333 0.33333333 0.33333333]
filter_var = np.ones(3)/3
print(filter_var)
'''
[100, 102, 98, 105, 103, 99, 101]
[1/3,1/3,1/3]
100 * 1/3 + 102 * 1/3 + 98 * 1/3
(100 + 102 + 98) * 1/3
300 * 1/3
100.0

[100, 102, 98, 105, 103, 99, 101]
102 * 1/3 + 98 * 1/3 + 105 * 1/3
(102 + 98 + 105) * 1/3
305 * 1/3
101.66666667
'''
def moving_average(data, window):
    return np.convolve(data, np.ones(window)/window, mode='valid')

# 함수 호출
ma_3 = moving_average(prices, 3)
print(ma_3)
```
