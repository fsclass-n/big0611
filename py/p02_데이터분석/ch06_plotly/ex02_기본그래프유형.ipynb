## 선 그래프(line plot)

```python
import plotly.express as px
import pandas as pd
import numpy as np
```

```python
num = np.random.randn(30)
num
```

```python
# 시계열 데이터 생성
# 2024-01-01 부터 D(Daily) 하루씩 30일간
dates = pd.date_range('2026-06-01', periods=30, freq='D')
# np.random.randn(30): 평균 0, 표준편차 1에 따른 난수 30개
values = np.cumsum(np.random.randn(30)) + 100

time_series = pd.DataFrame({
    '날짜': dates,
    '주가': values
})
time_series
```

```python
# 1. 단일 선 그래프
fig = px.line(
    time_series, 
    x='날짜', 
    y='주가',
    title='일일 주가 변동',
    markers="True"
)
fig.show()
```

```python
num = np.cumsum(np.random.randn(1))[0]
print(num)
# 참고: 숫자 범위 제한 없다.
```

```python
# 2. 다중 선 그래프
# 데이터 생성
companies = ['삼성전자', 'LG전자', 'SK하이닉스']
stock_data = []

for company in companies:
    for i, date in enumerate(dates):
        price = 100 + np.cumsum(np.random.randn(1))[0] * (i + 1) / 10
        stock_data.append({
            '날짜': date,
            '회사': company,
            '주가': price
        })

print(stock_data)

# 데이터 분석
df_stocks = pd.DataFrame(stock_data)
df_stocks
```

# 선 스타일

- dot

- dashdot
- longdashdot

```python
# 데이터 시각화
fig = px.line(
    df_stocks,
    x='날짜', # x축 컬럼명
    y='주가', # y축 컬럼명
    color='회사', # 그룹별 색상,
    title='회사별 주가 비교', # 그래프 제목
    line_dash='회사', # 선 스타일
    
    category_orders={
        '회사': ['SK하이닉스', 'LG전자', '삼성전자']
    },
    # color_discrete_sequence=['lightblue', 'orange', 'yellow'],
    color_discrete_map={
        '삼성전자': 'orange',
        'LG전자': 'yellow',
        'SK하이닉스': 'lime'
    },
    line_dash_map={
        '삼성전자': 'solid',
        'LG전자': 'longdashdot',
        'SK하이닉스': 'longdash'
    }
)
fig.show()
```

## 히스토그램(histogram)

```python
# 1. 기본 히스토그램
# 시험 점수 분포 데이터
# 데이터 생성
# NumPy의 난수만 고정
np.random.seed(42) 
# np.random.normal(평균, 표준편차, 개수)
# 평균 75, 표준편차 15인 정규분포(normal distribution)에서 랜덤 숫자 1000개를 생성
scores = np.random.normal(75, 15, 1000)
# 0~100 사이로 제한
# np.clip(배열, 최소값, 최대값)
scores = np.clip(scores, 0, 100)
scores
```

```python
# 데이터 분석
df_hist = pd.DataFrame({'점수': scores})
df_hist
```

```python
# 데이터 시각화
fig = px.histogram(
    df_hist, 
    x='점수',
    title='시험 점수 분포',
    nbins=20, # 구간 수
    marginal='rug' # 추가 통계 차트: rug, box, violin, histogram
)
'''
nbins=20: 데이터를 쪼갤 막대 격자(Bin)의 개수를 20개로 지정합니다. 0점부터 100점까지의 범위를 20개로 쪼갰으므로, 하나의 막대는 약 5점 단위(0~5, 5~10...)의 범위를 가지며 그 안에 속한 학생 수를 셉니다.
marginal='rug':히스토그램 상단(혹은 하단) 여백에 러그(Rug, 양탄자 술) 플롯을 추가합니다.데이터 포인트 1개당 아주 얇은 세로선 한 줄(|)을 그어주는 방식인데, 막대로 뭉뚱그려진 히스토그램 밑에 실제 데이터가 어디에 촘촘하게 몰려있고 어디가 텅 비었는지 미세하게 짚어주는 서브 차트 역할을 합니다.이 옵션에는 'rug' 외에도 박스플롯을 붙여주는 'box', 바이올린 플롯을 붙여주는 'violin' 등을 넣을 수 있습니다.
'''

fig.update_layout(
    xaxis_title="점수",
    yaxis_title="빈도"
)

fig.show()
```

# 2. 그룹별 히스토그램

```python
# 성별 키 분포 비교
height_data = []
np.random.seed(42)

# 남성 키 데이터(평균 175cm)
male_heights = np.random.normal(175, 6, 500)
for height in male_heights:
    height_data.append({'키': height, '성별': '남성'})

# 여성 키 데이터(평균 162cm)
female_heights = np.random.normal(162, 5, 500)
for height in female_heights:
    height_data.append({'키': height, '성별': '여성'})

df_heights = pd.DataFrame(height_data)
df_heights
```
