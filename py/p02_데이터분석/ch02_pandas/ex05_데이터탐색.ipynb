# 데이터 탐색 및 집계 연산

```python
import pandas as pd
```

```python
# 데이터 생성 (전자상거래 주문 데이터)
order_data = pd.DataFrame({
    'order_id': ['ORD001', 'ORD002', 'ORD003', 'ORD004', 'ORD005', 'ORD006', 'ORD007', 'ORD008'],
    'customer_id': ['C001', 'C002', 'C001', 'C003', 'C002', 'C004', 'C001', 'C003'],
    'product_category': ['전자제품', '의류', '전자제품', '도서', '의류', '전자제품', '도서', '의류'],
    'order_amount': [150000, 45000, 320000, 25000, 67000, 89000, 35000, 52000],
    'customer_age': [28, 35, 28, 42, 35, 31, 28, 42],
    'region': ['서울', '부산', '서울', '대구', '부산', '서울', '서울', '대구'],
    'order_date': pd.to_datetime(['2024-01-15', '2024-01-16', '2024-01-18', '2024-01-20',
                                  '2024-01-22', '2024-01-25', '2024-01-28', '2024-01-30'])    
})

# 출력
order_data
print(order_data)
display(order_data)
```

```python
order_data.to_csv('data.csv', index=False, encoding='utf-8-sig')
```

## 기술통계량 및 데이터 요약
- df.agg(['집계함수1', '집계함수2', ...]): aggregate(집계하다) 다중 집계 함수
    - 집계함수: 'count', 'sum', 'mean', 'median', 'std', 'var', 'min', 'max'
- df.quantile([0.25, 0.5, 0.75, ...]): 분위수 계산
- df.corr(): correlation(상관관계) 행렬
- df.nunique(): number of unique(고유 값 개수)
- df.mode(): 최빈값

```python
# 주문금액에 대한 다양한 통계량 계산
amount_stats = order_data['order_amount'].agg(['mean', 'median', 'std', 'min', 'max', 'count'])
print('=== 주문금액 통계량 ===')
print(amount_stats)
```

```python
# 분위수 계산
print('=== 주문금액 분위수 ===')
quantiles = order_data['order_amount'].quantile([0.25, 0.5, 0.75, 0.99])
print(quantiles)
```

```python
# 고객 연령과 주문금액 간 상관관계
print('=== 수치형 변수 간 상관관계 ===')
numeric_corr = order_data[['customer_age', 'order_amount']].corr()
print(numeric_corr)
```

```python
# 카테고리별 고유값 개수
print('=== 각 컬럼별 고유값 개수 ===')
print(order_data.nunique())
```

## 피봇 테이블(pivot_table)
### pd.pivot_table(df, values, index, columns, aggfunc, fill_value)
- df: 데이터프레임
- index: 행 인덱스 컬럼명
- columns: 열 인덱스 컬럼명
- aggfunc: 집계 함수
- values: 집계할 값의 컬럼명
- fill_value: 결측치(누락된 값)의 대체값

```python
# 지역별, 카테고리별 평균 주문금액 피봇테이블
region_category_pivot = pd.pivot_table(
    order_data,
    index='region',
    columns='product_category',
    aggfunc='mean',
    values='order_amount',
    fill_value=0
)
print('=== 지역별 카테고리별 평균 주문금액 ===')
print(region_category_pivot)
```

```python
# 다중 집계 함수 적용
# 지역별, 카테고리별 매출 평균, 건수, 합계
multi_agg_pivot = pd.pivot_table(
    order_data,
    index='region',
    columns='product_category',
    values='order_amount',
    aggfunc=['mean', 'count', 'sum'],
    fill_value=0,
    margins=True,  # 총합 추가
    margins_name='Total' # 총합 행 이름
)
print('=== 다중 집계 함수 피봇테이블 (일부) ===')
print(multi_agg_pivot['mean'])
```

```python
# 고객 연령대별, 지역별 주문 건수
order_data['age_group'] = pd.cut(
    order_data['customer_age'], 
    bins=[0, 30, 40, 50, 100],
    labels=['20-30대', '30-40대', '40-50대', '50대 이상']
)
order_data
```

```python
# 연령대별, 지역별, 주문 건수
age_region_pivot = pd.pivot_table(
    order_data, 
    index='age_group', 
    columns='region', 
    values='order_id', 
    aggfunc='count',
    fill_value=0
)
print('=== 연령대별 지역별 주문 건수 ===')
print(age_region_pivot)
```

## 그룹별 집계 연산(groupby)

```python
# 고객별 주문 통계
customer_stats = order_data.groupby('customer_id').agg({
    'order_amount': ['count','sum','mean', 'max'],
    'order_id': 'count'
}).round(2)

display(customer_stats)
```

```python
# 지역별 상세 분석
region_analysis = order_data.groupby('region').agg({
    'order_amount': ['count', 'sum', 'mean', 'std'],
    'customer_age': 'mean',
    'customer_id': 'nunique'
}).round(2)
display(region_analysis)
```

```python
# 상품 카테고리별 월별 주문 트렌드
order_data['order_month'] = order_data['order_date'].dt.month
order_data
```

```python
monthly_category = order_data.groupby(['order_month', 'product_category']).agg({
    'order_amount': ['count', 'sum'],
    'customer_id': 'nunique'
})
print('=== 월별 카테고리별 주문 현황 ===')
print(monthly_category)
```

```python
# 그룹별 크기와 기술통계
print('=== 지역별 그룹 크기 ===')
print(order_data.groupby('region').size())
```

```python
# 카테고리별 주문금액 기술통계
print('=== 카테고리별 주문금액 기술통계 ===')
category_describe = order_data.groupby('product_category')['order_amount'].describe()
print(category_describe.round(2))
```
